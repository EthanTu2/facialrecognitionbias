## Why are we studying this?

Facial recognition systems are often described as purely technical tools, but they also have real social consequences.

When people first learn computer science, it is easy to focus only on whether code runs correctly. However, in practice, technical decisions can shape social outcomes. For example:
- what data is collected
- which labels are used
- who is represented in the dataset
- what metric is chosen for evaluation

These decisions may seem technical, but they influence who benefits from a system and who is harmed by its mistakes.

Importantly, even if a model is advanced, if the data is biased, it can still produce highly biased results. This means that a system that appears technically successful may still behave unevenly across different groups.

In this tutorial, we will learn how to measure and analyze bias in facial recognition systems by examining differences in performance across demographic groups and how these differences are influenced by the data.

### Key Terms

Here are a few terms that will appear throughout the tutorial.

- **Prediction**: the output produced by a model
- **Ground truth**: the correct label or answer
- **Error**: a case where the prediction does not match the ground truth
- **Bias**: uneven performance across demographic groups, or systematic error from flawed model assumptions
- **Demographic group**: a category such as race or gender presentation used to examine whether performance differs across populations
