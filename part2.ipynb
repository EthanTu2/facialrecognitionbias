## Part 2: Understanding the Data and Model

Before we begin analyzing performance, we first need to understand what a facial recognition system is and what kind of data it requires.

### What is a facial recognition model?

A facial recognition model is a machine learning system that takes an image of a face as input and produces a prediction about identity.

At a high level, the system:
- receives an image
- processes visual features of the face
- compares those features to known examples
- outputs a prediction

Although this process may seem straightforward, the way the model represents and compares faces plays a critical role in its behavior.

### Where does the data come from?

The data used in this tutorial is derived from publicly available facial image datasets on Kaggle (https://www.kaggle.com/datasets/aibloy/fairface).
![Example faces](fairface.png)

These datasets contain:
- images of human faces
- identity labels
- demographic attributes (age, gender, and race)

These attributes allow us to examine how model performance may differ across groups.

### From images to embeddings

Instead of working directly with raw images, facial recognition systems typically convert each image into a numerical representation called an **embedding**.

An embedding is a vector (a list of numbers) that captures important features of a face, such as:
- shape
- structure
- distinguishing characteristics

This allows the model to compare faces mathematically rather than visually. Since working with embeddings makes the problem faster (no need to process images each time), simpler (we work with numbers instead of pixels), and more consistent for comparison, we often use embeddings for facial recognition models.

In our case, we generated these embeddings using the DeepFace library with a pre-trained FaceNet model. This model converts each face image into a numerical vector that represents its key features.

In this tutorial, we use precomputed embeddings so that we can focus on analyzing model behavior rather than building the embedding model itself.

### What comes next

Now that we understand what the model does and how the data is represented, we can begin examining its predictions and measuring its performance.
